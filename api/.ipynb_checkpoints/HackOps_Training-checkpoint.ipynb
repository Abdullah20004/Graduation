# 🎯 HackOps Multi-Agent Adversarial Training

Train Blue (Defender) and Red (Attacker) agents together in an adversarial scenario.

**Features:**
- 🔄 Multi-agent adversarial training
- 📊 Live training visualization
- ⚙️ Easy configuration with sliders
- 💾 Automatic model saving
- 📈 Real-time metrics (loss, rewards, episode length)

## 1. Setup and Imports

```python
import sys
import io
import os
from pathlib import Path
import time
from datetime import datetime
import json

# Fix encoding for Windows
if sys.platform == 'win32':
    sys.stdout = io.TextIOWrapper(sys.stdout.buffer, encoding='utf-8')
    sys.stderr = io.TextIOWrapper(sys.stderr.buffer, encoding='utf-8')

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import display, clear_output, HTML
import ipywidgets as widgets
import plotly.graph_objects as go
from plotly.subplots import make_subplots

from stable_baselines3 import PPO, A2C
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.callbacks import BaseCallback
from stable_baselines3.common.monitor import Monitor

from train_sb3 import create_training_env, GymToGymnasiumWrapper

# Enable interactive plotting
%matplotlib inline
plt.style.use('seaborn-v0_8-darkgrid')

print("✅ All imports successful!")
```

## 2. Training Configuration

Adjust training parameters using sliders:

```python
# Training configuration widgets
timesteps_widget = widgets.IntSlider(
    value=50000,
    min=5000,
    max=500000,
    step=5000,
    description='Timesteps:',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='500px')
)

lr_widget = widgets.FloatLogSlider(
    value=0.0003,
    base=10,
    min=-5,
    max=-2,
    step=0.1,
    description='Learning Rate:',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='500px')
)

algorithm_widget = widgets.Dropdown(
    options=['PPO', 'A2C'],
    value='PPO',
    description='Algorithm:',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='300px')
)

train_mode_widget = widgets.RadioButtons(
    options=[
        ('Train Blue Only', 'blue'),
        ('Train Red Only', 'red'),
        ('Train Both (Adversarial)', 'adversarial')
    ],
    value='adversarial',
    description='Training Mode:',
    style={'description_width': '120px'}
)

display(widgets.VBox([
    widgets.HTML("<h3>⚙️ Training Configuration</h3>"),
    train_mode_widget,
    timesteps_widget,
    lr_widget,
    algorithm_widget
]))

print("\n✅ Configuration ready! Scroll down to start training.")
```

## 3. Live Training Visualization

Custom callback for real-time plotting:

```python
class LivePlottingCallback(BaseCallback):
    """Callback for live training visualization"""
    
    def __init__(self, agent_name='Agent', plot_freq=2048, verbose=0):
        super().__init__(verbose)
        self.agent_name = agent_name
        self.plot_freq = plot_freq
        
        # Metrics storage
        self.timesteps = []
        self.rewards = []
        self.ep_lengths = []
        self.losses = []
        self.policy_losses = []
        self.value_losses = []
        self.entropies = []
        
        # Setup figure
        self.fig = None
        self.setup_figure()
    
    def setup_figure(self):
        """Create the plotting figure"""
        self.fig = make_subplots(
            rows=2, cols=2,
            subplot_titles=(
                f'{self.agent_name} - Mean Reward',
                f'{self.agent_name} - Episode Length',
                f'{self.agent_name} - Policy Loss',
                f'{self.agent_name} - Value Loss'
            ),
            vertical_spacing=0.12,
            horizontal_spacing=0.1
        )
        
        self.fig.update_layout(
            height=600,
            showlegend=False,
            title_text=f"📊 Live Training Metrics - {self.agent_name}",
            title_x=0.5
        )
    
    def _on_step(self) -> bool:
        if self.n_calls % self.plot_freq == 0:
            # Collect metrics
            self.timesteps.append(self.num_timesteps)
            
            if len(self.model.ep_info_buffer) > 0:
                self.rewards.append(np.mean([ep['r'] for ep in self.model.ep_info_buffer]))
                self.ep_lengths.append(np.mean([ep['l'] for ep in self.model.ep_info_buffer]))
            
            # Get training stats
            if hasattr(self.locals, 'loss'):
                self.losses.append(self.locals.get('loss', 0))
            if hasattr(self.locals, 'pg_loss'):
                self.policy_losses.append(self.locals.get('pg_loss', 0))
            if hasattr(self.locals, 'value_loss'):
                self.value_losses.append(self.locals.get('value_loss', 0))
            if hasattr(self.locals, 'entropy_loss'):
                self.entropies.append(self.locals.get('entropy_loss', 0))
            
            # Update plots
            self.update_plots()
        
        return True
    
    def update_plots(self):
        """Update the live plots"""
        clear_output(wait=True)
        
        # Clear existing traces
        self.fig.data = []
        
        # Plot rewards
        if self.rewards:
            self.fig.add_trace(
                go.Scatter(x=self.timesteps, y=self.rewards, mode='lines', name='Reward',
                          line=dict(color='green', width=2)),
                row=1, col=1
            )
        
        # Plot episode lengths
        if self.ep_lengths:
            self.fig.add_trace(
                go.Scatter(x=self.timesteps, y=self.ep_lengths, mode='lines', name='Ep Length',
                          line=dict(color='blue', width=2)),
                row=1, col=2
            )
        
        # Plot policy loss
        if self.policy_losses:
            self.fig.add_trace(
                go.Scatter(x=self.timesteps, y=self.policy_losses, mode='lines', name='Policy Loss',
                          line=dict(color='red', width=2)),
                row=2, col=1
            )
        
        # Plot value loss
        if self.value_losses:
            self.fig.add_trace(
                go.Scatter(x=self.timesteps, y=self.value_losses, mode='lines', name='Value Loss',
                          line=dict(color='orange', width=2)),
                row=2, col=2
            )
        
        # Update axes
        self.fig.update_xaxes(title_text="Timesteps", row=2, col=1)
        self.fig.update_xaxes(title_text="Timesteps", row=2, col=2)
        
        display(self.fig)
        
        # Show current stats
        if self.rewards:
            print(f"\n📈 Current Stats (Timestep {self.num_timesteps}):")
            print(f"   Reward: {self.rewards[-1]:.2f}")
            if self.ep_lengths:
                print(f"   Episode Length: {self.ep_lengths[-1]:.1f}")
            if self.policy_losses:
                print(f"   Policy Loss: {self.policy_losses[-1]:.6f}")

print("✅ Live plotting callback created!")
```

## 4. Training Functions

```python
def train_single_agent(agent_type, timesteps, learning_rate, algorithm='PPO'):
    """Train a single agent with live visualization"""
    print(f"\n{'='*70}")
    print(f"🎯 Training {agent_type.upper()} Agent")
    print(f"{'='*70}")
    print(f"Algorithm: {algorithm}")
    print(f"Timesteps: {timesteps:,}")
    print(f"Learning Rate: {learning_rate}")
    print(f"{'='*70}\n")
    
    # Create environment
    print("🌐 Creating environment...")
    env = create_training_env(agent_type)
    print("✅ Environment ready\n")
    
    # Create model
    print(f"🏗️ Building {algorithm} model...")
    if algorithm == 'PPO':
        model = PPO(
            'MlpPolicy',
            env,
            learning_rate=learning_rate,
            n_steps=2048,
            batch_size=64,
            n_epochs=10,
            gamma=0.99,
            verbose=0
        )
    else:  # A2C
        model = A2C(
            'MlpPolicy',
            env,
            learning_rate=learning_rate,
            gamma=0.99,
            verbose=0
        )
    print("✅ Model created\n")
    
    # Create callback
    callback = LivePlottingCallback(agent_name=f"{agent_type.upper()} Agent")
    
    # Train
    print("🎓 Starting training...")
    print("📊 Live plots will appear below:\n")
    
    start_time = time.time()
    model.learn(total_timesteps=timesteps, callback=callback)
    duration = time.time() - start_time
    
    # Save model
    model_dir = Path('models')
    model_dir.mkdir(exist_ok=True)
    model_path = model_dir / f'{agent_type}_agent_notebook.zip'
    model.save(model_path)
    
    print(f"\n{'='*70}")
    print("✅ TRAINING COMPLETE")
    print(f"{'='*70}")
    print(f"Duration: {duration:.1f} seconds")
    print(f"Model saved: {model_path}")
    print(f"{'='*70}\n")
    
    env.close()
    return model, callback

def train_adversarial(timesteps, learning_rate, algorithm='PPO'):
    """Train both agents in adversarial scenario"""
    print(f"\n{'='*70}")
    print("⚔️  ADVERSARIAL MULTI-AGENT TRAINING")
    print(f"{'='*70}")
    print(f"Algorithm: {algorithm}")
    print(f"Timesteps per agent: {timesteps:,}")
    print(f"Learning Rate: {learning_rate}")
    print(f"{'='*70}\n")
    
    # Train agents alternately
    half_timesteps = timesteps // 2
    
    print("\n🔵 Phase 1: Training BLUE (Defender) Agent")
    print("="*70)
    blue_model, blue_callback = train_single_agent('blue', half_timesteps, learning_rate, algorithm)
    
    print("\n🔴 Phase 2: Training RED (Attacker) Agent")
    print("="*70)
    red_model, red_callback = train_single_agent('red', half_timesteps, learning_rate, algorithm)
    
    print("\n" + "="*70)
    print("✅ ADVERSARIAL TRAINING COMPLETE")
    print("="*70)
    print("Both agents trained successfully!")
    print("Models saved in ./models/")
    print("="*70 + "\n")
    
    return blue_model, red_model, blue_callback, red_callback

print("✅ Training functions ready!")
```

## 5. START TRAINING

Click the button below to start training with your configured parameters:

```python
# Training status
status_output = widgets.Output()

def start_training(b):
    with status_output:
        clear_output()
        
        # Get configuration
        mode = train_mode_widget.value
        timesteps = timesteps_widget.value
        lr = lr_widget.value
        algorithm = algorithm_widget.value
        
        try:
            if mode == 'adversarial':
                train_adversarial(timesteps, lr, algorithm)
            else:
                train_single_agent(mode, timesteps, lr, algorithm)
        except Exception as e:
            print(f"\n❌ Error during training: {e}")
            import traceback
            traceback.print_exc()

# Start button
start_button = widgets.Button(
    description='🚀 START TRAINING',
    button_style='success',
    layout=widgets.Layout(width='300px', height='50px'),
    style={'font_weight': 'bold'}
)
start_button.on_click(start_training)

display(widgets.VBox([
    start_button,
    status_output
]))
```

## 6. Evaluate Trained Models

After training, evaluate your models:

```python
def evaluate_model(agent_type='blue', n_episodes=10):
    """Evaluate a trained model"""
    model_path = Path(f'models/{agent_type}_agent_notebook.zip')
    
    if not model_path.exists():
        print(f"❌ Model not found: {model_path}")
        print("Please train the model first.")
        return
    
    print(f"\n{'='*70}")
    print(f"📈 Evaluating {agent_type.upper()} Agent")
    print(f"{'='*70}\n")
    
    # Load model
    if 'PPO' in str(model_path) or True:  # Assume PPO for now
        model = PPO.load(model_path)
    
    # Create environment
    env = create_training_env(agent_type)
    
    # Evaluate
    print(f"Running {n_episodes} evaluation episodes...\n")
    mean_reward, std_reward = evaluate_policy(
        model, env, n_eval_episodes=n_episodes, return_episode_rewards=False
    )
    
    print(f"{'='*70}")
    print("📊 Evaluation Results")
    print(f"{'='*70}")
    print(f"Episodes: {n_episodes}")
    print(f"Mean Reward: {mean_reward:.2f}")
    print(f"Std Reward: {std_reward:.2f}")
    print(f"{'='*70}\n")
    
    env.close()
    return mean_reward, std_reward

# Evaluation widgets
eval_agent = widgets.Dropdown(
    options=['blue', 'red'],
    description='Agent:',
    style={'description_width': '100px'}
)

eval_episodes = widgets.IntSlider(
    value=10,
    min=1,
    max=100,
    description='Episodes:',
    style={'description_width': '100px'}
)

eval_output = widgets.Output()

def run_evaluation(b):
    with eval_output:
        clear_output()
        evaluate_model(eval_agent.value, eval_episodes.value)

eval_button = widgets.Button(
    description='📊 Evaluate Model',
    button_style='info',
    layout=widgets.Layout(width='200px')
)
eval_button.on_click(run_evaluation)

display(widgets.VBox([
    widgets.HTML("<h3>📊 Model Evaluation</h3>"),
    eval_agent,
    eval_episodes,
    eval_button,
    eval_output
]))
```

## 7. Compare Agent Performance

```python
def compare_agents(n_episodes=10):
    """Compare Blue vs Red agent performance"""
    print("\n" + "="*70)
    print("⚔️  AGENT COMPARISON")
    print("="*70 + "\n")
    
    # Evaluate both
    blue_reward, blue_std = evaluate_model('blue', n_episodes)
    red_reward, red_std = evaluate_model('red', n_episodes)
    
    # Create comparison plot
    fig = go.Figure(data=[
        go.Bar(
            x=['Blue (Defender)', 'Red (Attacker)'],
            y=[blue_reward, red_reward],
            error_y=dict(type='data', array=[blue_std, red_std]),
            marker_color=['blue', 'red']
        )
    ])
    
    fig.update_layout(
        title="Agent Performance Comparison",
        yaxis_title="Mean Reward",
        showlegend=False,
        height=400
    )
    
    display(fig)

compare_button = widgets.Button(
    description='⚔️ Compare Agents',
    button_style='warning',
    layout=widgets.Layout(width='200px')
)

compare_output = widgets.Output()

def run_comparison(b):
    with compare_output:
        clear_output()
        compare_agents(10)

compare_button.on_click(run_comparison)

display(widgets.VBox([
    compare_button,
    compare_output
]))
```
